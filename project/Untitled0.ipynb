```python
import csv
```

```python
f = open('melanoma.csv', 'rb')
reader = csv.reader(f)
rownum = 0
year = []
incidence = []
for row in reader:
    if  rownum == 0:
        header = row
    else:
        year.append(row[0])
        incidence.append(row[1])
    rownum += 1
r=map(float,year)
n=map(float,incidence)    
print r
print n
f.close()
```

```python
from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('melanoma year and incidence')
ax.set_xlabel('year')
ax.set_ylabel('incidence')
ax.plot(r,n)
fig.savefig("oyster_with_titles.png")
```

this graph describes how many melanoma incidents happend in every year. and we can see that the numbers are getting higher
